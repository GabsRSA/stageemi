```python
import stageemi
import stageemi.dev.visu as dev
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import stageemi.dev.decorator_map as dm
import ipywidgets as widg 
import ipyleaflet as ipyl
import datetime as dt
import os 
from ipywidgets import Text, HTML
from ipyleaflet import WidgetControl
import pandas 
```

```python
file_compas= '../utils/distance_compas.csv'
df_compas = pandas.read_csv(file_compas,sep=',')
# df_compas
```

```python
penalty_dict={"9-10":4,"7-8":3,"4-6":2,"0-3":1} # needs to be done in reverse order (high values to low values)
penalty_dict
```

```python
df_compas.columns
```

```python
N=len(df_compas.index)
ind={"0-3":np.nan,"4-6":np.nan,"7-8":np.nan,"9-10":np.nan}

for i,val in enumerate(df_compas.columns[1:19]): # goes till column 19-1=18
    #i=2 in debug case when one does not start from column 1

    for j,key in enumerate(penalty_dict):  
        if val != '18':
            ind[key] = np.where(np.all(np.array((df_compas[val][i+1::]>=int(key.split('-')[0]),df_compas[val][i+1::]<=int(key.split('-')[1]))),axis=0))[0]                  
            if len(ind[key])>0:        
                df_compas[val][ind[key]+i+1]=df_compas[val][ind[key]+i+1]+penalty_dict[key]
        elif val == '18': # special treatment for the last but one column
            if np.all(np.array((list(df_compas[val][i+1::]>=int(key.split('-')[0])),list(df_compas[val][i+1::]<=int(key.split('-')[1]))))):
                df_compas[val][N-1]=df_compas[val][N-1]+penalty_dict[key]                    
```
