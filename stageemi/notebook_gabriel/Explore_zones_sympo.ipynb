```python
zone_dpt="41"
fname_mask = '../GeoData/zones_sympo_multiples/'+zone_dpt+'_mask_zones_sympos.nc'
da_mask = xr.open_dataarray(fname_mask)
```

```python
zs_l=[zs for zs in da_mask.id.values.tolist() if "+" not in zs]
zs_N=len(zs_l)
```

```python
fig,axes = plt.subplots(nrows=int(np.ceil(zs_N/2)),ncols = 2,figsize=(14,15))
ax = axes.flat

for izs, zs in enumerate(zs_l):
    da_mask.isel(id=izs).plot.imshow(ax=ax[izs])
    ax[izs].set_title(zs,fontsize=15)
plt.tight_layout()
```
