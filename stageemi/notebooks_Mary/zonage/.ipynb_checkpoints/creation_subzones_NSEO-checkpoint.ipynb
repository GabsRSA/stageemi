```python
import numpy as np
import glob
import xarray as xr 
import matplotlib.pyplot as plt
from collections import namedtuple
import pandas
```

```python
def get_index(data,val):
    return np.argmin(np.abs(data-val))

def read_xarray(file): 
    data = xr.open_dataset(file)
    data["latitude"]  = data["latitude"].round(5)
    data["longitude"] = data["longitude"].round(5)
    data.close()
    return data


def conversion(ds):
    """convert wwmf code into wme code"""
    file_CodesWWMF= '../../utils/CodesWWMF.csv'
    df_WWMF = pandas.read_csv(file_CodesWWMF,usecols = (0,1,6,7),sep=',')
    
    ds["wme_arr"]=ds.unknown

    for iwwmf,wwmf in enumerate(df_WWMF["Code WWMF"]):
        ds["wme_arr"]=ds["wme_arr"].where(ds.unknown!=wwmf,df_WWMF["Code WME"][iwwmf])
        
    return ds

Subzones = namedtuple('Subzones',['lat1','lat2','lon1','lon2'])
```

```python
dir_mask = '/home/mrpa/borderiesm/stageEMI/Codes/stageemi/stageemi/GeoData/nc_departement/'
dep = 'FRK24'
dep_file  = dir_mask + dep +'.nc'
ds_dep    = read_xarray(dep_file) 
ds_mask = ds_dep.copy().squeeze("id") # on va creer un dataset qui contient les mask N, S, E, O
latmin_dict = {'nord':float(lat2_3.values),'sud':ds_mask.latitude.values.min()
               ,'ouest':ds_mask.latitude.values.min(),'est':ds_mask.latitude.values.min()}

latmax_dict = {'nord':ds_mask.latitude.values.max(),'sud':float(lat1_3.values)
               ,'ouest':ds_mask.latitude.values.max(),'est':ds_mask.latitude.values.max()}

lonmin_dict = {'nord':ds_mask.longitude.values.min(),'sud':ds_mask.longitude.values.min()
               ,'ouest':ds_mask.longitude.values.min(),'est':float(lon2_3.values)}

lonmax_dict = {'nord':ds_mask.longitude.values.max(),'sud':ds_mask.longitude.values.max()
               ,'ouest':float(lon1_3.values),'est':ds_mask.longitude.values.max()}

fig,axes = plt.subplots(nrows=2,ncols =2)
ax = axes.flat
for imask_id,mask_id in enumerate(['nord','sud','est','ouest']):
    ds_mask[mask_id] = ds_mask.mask * 0
    ind    = np.where((ds_mask.latitude>=latmin_dict[mask_id]) & (ds_mask.latitude<=latmax_dict[mask_id])
             & (ds_mask.longitude >= lonmin_dict[mask_id]) & (ds_mask.longitude<=lonmax_dict[mask_id]))
    ds_mask[mask_id].values[ind] = 1
    ds_mask[mask_id].values[np.isnan(ds_dep.squeeze("id").mask.values)] = np.nan
    ds_mask[mask_id].plot.imshow(ax=ax[imask_id])
    ax[imask_id].set_title(mask_id)
fig.tight_layout()


```

```python
# defintion des zones 
lst_zones = {}

lst_zones['centre-centre']    = Subzones(lat1=float(lat1_3.values),lat2=float(lat2_3.values),
                               lon1=float(lon1_3.values),lon2=float(lon2_3.values))

lst_zones['est-centre']       = Subzones(lat1=float(lat1_3.values),lat2=float(lat2_3.values),
                               lon1=float(lon2_3.values),lon2=ds_Isere.longitude.values.max())

lst_zones['ouest-centre']      = Subzones(lat1=float(lat1_3.values),lat2=float(lat2_3.values),
                               lon1=ds_Isere.longitude.values.min(),lon2=float(lon1_3.values))

lst_zones['sud-centre']        = Subzones(lat1=ds_Isere.latitude.values.min(),lat2=float(lat1_3.values),
                               lon1=float(lon1_3.values),lon2=float(lon2_3.values))

lst_zones['nord-centre'] = Subzones(lat1=float(lat2_3.values),lat2=ds_Isere.latitude.values.max(),
                               lon1=float(lon1_3.values),lon2=float(lon2_3.values))

lst_zones['sud-est']    = Subzones(lat1=ds_Isere.latitude.values.min(),lat2=float(lat1_3.values),
                               lon1=float(lon2_3.values),lon2=ds_Isere.longitude.values.max())

lst_zones['sud-ouest']  = Subzones(lat1=ds_Isere.latitude.values.min(),lat2=float(lat1_3.values),
                               lon1=ds_Isere.longitude.values.min(),lon2=float(lon1_3.values))

lst_zones['nord-est']   = Subzones(lat1=float(lat2_3.values),lat2=ds_Isere.latitude.values.max(),
                               lon1=float(lon2_3.values),lon2=ds_Isere.longitude.values.max())

lst_zones['nord-ouest'] = Subzones(lat1=float(lat2_3.values),lat2=ds_Isere.latitude.values.max(),
                               lon1=ds_Isere.longitude.values.min(),lon2=float(lon1_3.values))

# zones verticales (est,  ouest, centre-h), seules les latitudes varient pour les dfinir
for var in ['est','ouest','centre']:
    if var == 'centre':
        lst_zones['centre-h'] = Subzones(lat1 = lst_zones["sud-"+var].lat1, lat2 = lst_zones["nord-"+var].lat2, 
                lon1 = lst_zones["sud-"+var].lon1, lon2 = lst_zones["sud-"+var].lon2)
    else: 
        lst_zones[var] = Subzones(lat1 = lst_zones["sud-"+var].lat1, lat2 = lst_zones["nord-"+var].lat2, 
                lon1 = lst_zones["sud-"+var].lon1, lon2 = lst_zones["sud-"+var].lon2)
        
# zones horizontales (Nord, sud, centre-v) # seules les longitudes varient pour dfinir chacune des zones
for var in ['nord','sud','centre']:
    if var == 'centre':
        lst_zones['centre-v'] = Subzones(lat1 = lst_zones["centre-"+var].lat1, lat2 = lst_zones["centre-"+var].lat2, 
                lon1 = lst_zones["ouest-"+var].lon1, lon2 = lst_zones["est-"+var].lon2)
    else: 
        lst_zones[var] = Subzones(lat1 = lst_zones[var+"-centre"].lat1, lat2 = lst_zones[var+"centre-"].lat2, 
                lon1 = lst_zones[var+"-ouest"].lon1, lon2 = lst_zones[var+"-est"].lon2)
```

```python
file_CodesWWMF = '../../utils/CodesWWMF.csv'
df             = pandas.read_csv(file_CodesWWMF,usecols = (0,1,6,7),sep=',')
Code_WME       = df['Code WME'].to_numpy()
Code_WWMF      = df['Code WWMF'].to_numpy()
legende_WWMF   = df['Legende WWMF'].to_numpy()
legende_WME    = df['Legende WME'].to_numpy()

cible_list  = np.unique(ds_Isere['wme_arr'].values[~np.isnan(ds_Isere['wme_arr'])])# on parcourt les temps sensibles disponibles sur la zone
legend_list = [np.unique(legende_WME[Code_WME == WME])[0] for WME in cible_list] 
legend_list
```

```python
zones_impactees = {}
for icible,cible in enumerate(cible_list):
    print(legend_list[icible],cible)
    zones_impactees[cible] = {}
    for zone in lst_zones.keys():
        var_zone = ds_Isere.where((ds_Isere.latitude>=lst_zones[zone].lat1) & (ds_Isere.latitude<lst_zones[zone].lat2)
                         & (ds_Isere.longitude>=lst_zones[zone].lon1) & (ds_Isere.longitude<lst_zones[zone].lon2)
             ,drop=True)

        somme = np.sum(var_zone["wme_arr"].values == cible)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            
        nombre_mailles = var_zone["wme_arr"].values.size
        fraction = somme / nombre_mailles
       
        if fraction >0.05:
            zones_impactees[cible][zone] = fraction
            print(zone, fraction)
    print('')     
```
